```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
# color, marker, linewidth, labels for axis, datasets and the title, axis, grid, text
#x = np.array([1,2,3,4])
x = np.arange(0, 5, 0.1)
y  = x **3
y2 = x**2
# plt.plot(x, y, color = "black", marker="p")
plt.plot(x, y, color = "black", linewidth = 5, label = "x^3")

plt.plot(x, y2, color = "red", linewidth = 5, label = "x^2")

# plt.axis([0, 10, 0, 200]) # using this x axis is taken from 0 to 10 and y axis is taken from 0 to 200

plt.ylabel("y")
plt.xlabel("x")

plt.grid()

plt.text(2, 80, "Text", fontsize=12) # to add something at someplace in graph

plt.legend() # it adds the labels specified in plot() to graph, without it lables(x^3,x^2) will not show on graph
plt.title("Matplotlib demo")
plt.show()
```
