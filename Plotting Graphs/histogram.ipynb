```python
b=[1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14]
```

```python
import numpy as np
```

```python
xt=np.arange(23)
xt
```

```python
plt.hist(b,edgecolor="black")
plt.xticks(xt) # In order to see which numbers are clubed together
plt.show() # it shows the combined frequency of 1,2 3 as (2+1+3=6) 
# There are 10 bins , size of each bin is calculated by using(max-min)/bin i.e. (22-1)/10=2.1 this is how it is calculated
```

```python
plt.hist(b,bins=25,edgecolor="black")
plt.xticks(xt) 
plt.show() 
```

```python
plt.hist(b,bins=21,edgecolor="black")
plt.xticks(xt) 
plt.show()
# from bin 1 to bin 20
#bin1 [1,2) only 1 is include 2 is not
#bin2 [2,3)---------------only in last bin [21,22](both 21 znd 22 is included)
```

```python
plt.hist(b,bins=[1,3,6,10,15,20],edgecolor="black")# [1,3),[3,6),[6,10),[10,15),[15,20)
plt.show()
```

```python
plt.hist(b,bins=[1,3,6,10,15,20],edgecolor="black")# [1,3),[3,6),[6,10),[10,15),[15,20)
plt.axis([0,25,0,10])
plt.show()
```
