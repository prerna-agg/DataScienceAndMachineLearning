```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [2,4,6]

plt.scatter(x, y) #scatters the points all over the graph
plt.show() #only on calling show() graph will show up
# it picks a random color for us to show graph
```

```python
# to connect the points and make graph use plot() it doesnot show the points
x = [1,2,3]
y = [2,4,6]

x2 = [3,4,5]
y2 = [2,4,6]

plt.plot(x2, y2, "b-.") #"b-."--> first character in this string denotes the colour, 2nd and 2nd onwards character(-.) shows how you want your line to show up.These are some fixed values

plt.plot(x, y)
plt.show()
```

```python
# plot function x^3
import numpy as np
#x = np.array([1,2,3,4])
x = np.arange(0, 5, 0.1) # give starting point, ending point, step to get an array 0,0.1----4.9
y  = x **3 # its very easy to cube in a numpy array
plt.plot(x, y, "b")
plt.show()
```
